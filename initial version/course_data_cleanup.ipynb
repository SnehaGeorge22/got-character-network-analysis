```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph with signed edges
G = nx.DiGraph()

# Add edges with signs (+1 for alliances, -1 for conflicts)
edges = [
    ("Jon Snow", "Daenerys", 1),
    ("Jon Snow", "Night King", -1),
    ("Cersei", "Tyrion", -1),
    ("Arya", "The Hound", 1),
    ("Sansa", "Littlefinger", -1),
    ("Jaime", "Brienne", 1),
    ("Cersei", "Euron", 1)
]

# Add edges to graph
for u, v, sign in edges:
    G.add_edge(u, v, weight=sign)

# Visualization with different colors for positive and negative relationships
plt.figure(figsize=(8,6))
pos = nx.spring_layout(G, seed=42)

# Draw nodes
nx.draw(G, pos, with_labels=True, node_color="lightblue", edge_color="gray", node_size=3000, font_size=10, font_weight="bold")

# Draw positive edges in blue and negative edges in red
positive_edges = [(u, v) for u, v, d in G.edges(data=True) if d["weight"] == 1]
negative_edges = [(u, v) for u, v, d in G.edges(data=True) if d["weight"] == -1]

nx.draw_networkx_edges(G, pos, edgelist=positive_edges, edge_color="blue", width=2)
nx.draw_networkx_edges(G, pos, edgelist=negative_edges, edge_color="red", width=2, style="dashed")

plt.title("Signed Network of Game of Thrones Characters")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import time

# Define network changes over seasons
seasons = {
    1: [("Ned Stark", "Robert Baratheon", 1), ("Ned Stark", "Cersei Lannister", -1)],
    2: [("Stannis Baratheon", "Renly Baratheon", -1)],
    3: [("Jaime Lannister", "Brienne of Tarth", 1)],
    5: [("Jon Snow", "Daenerys Targaryen", 1)],
    8: [("Jon Snow", "Daenerys Targaryen", -1)]
}

# Function to visualize network for a given season
def plot_network(season):
    G = nx.Graph()
    
    # Add edges for the current and previous seasons
    for s in range(1, season+1):
        if s in seasons:
            for u, v, sign in seasons[s]:
                G.add_edge(u, v, weight=sign)
    
    pos = nx.spring_layout(G, seed=42)

    # Draw nodes
    nx.draw(G, pos, with_labels=True, node_color="lightblue", edge_color="gray", node_size=3000, font_size=10, font_weight="bold")

    # Draw edges with color coding
    positive_edges = [(u, v) for u, v, d in G.edges(data=True) if d["weight"] == 1]
    negative_edges = [(u, v) for u, v, d in G.edges(data=True) if d["weight"] == -1]

    nx.draw_networkx_edges(G, pos, edgelist=positive_edges, edge_color="blue", width=2)
    nx.draw_networkx_edges(G, pos, edgelist=negative_edges, edge_color="red", width=2, style="dashed")

    plt.title(f"Game of Thrones Dynamic Network - Season {season}")
    plt.show()

# Animate through seasons
for season in seasons.keys():
    plot_network(season)
    time.sleep(1)
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()

# Add conflict edges (Winner -> Loser)
conflicts = [
    ("Lannisters", "Starks", "win"),
    ("Jon Snow", "Ramsay Bolton", "win"),
    ("Daenerys", "Cersei", "win"),
    ("Jon Snow", "Daenerys", "betray")
]

# Define edge colors based on conflict result
edge_colors = {
    "win": "green",        # Victorious side
    "lose": "red",         # Defeated side
    "betray": "black"      # Betrayal or unexpected conflict
}

# Add nodes and edges with attributes
for winner, loser, result in conflicts:
    G.add_edge(winner, loser, color=edge_colors[result])

# Positioning of nodes
pos = nx.spring_layout(G, seed=42)

# Extract edge colors
edges = G.edges(data=True)
edge_colors = [data["color"] for _, _, data in edges]

# Draw nodes with different sizes for winners (bigger) and losers (smaller)
node_sizes = {node: 3000 if node in [w for w, _, _ in conflicts] else 1500 for node in G.nodes}
nx.draw(G, pos, with_labels=True, node_size=[node_sizes[n] for n in G.nodes], font_size=10, font_weight="bold", edge_color=edge_colors)

# Show the graph
plt.title("Game of Thrones Conflict Network")
plt.show()
```

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import vaderSentiment
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

# Sample dataset for character interactions over seasons
data = [
    {"season": 1, "char_a": "Ned Stark", "char_b": "Robert Baratheon", "interaction": "alliance", "sentiment": 0.7},
    {"season": 1, "char_a": "Ned Stark", "char_b": "Cersei Lannister", "interaction": "conflict", "sentiment": -0.8},
    {"season": 3, "char_a": "Jaime Lannister", "char_b": "Brienne of Tarth", "interaction": "trust", "sentiment": 0.5},
    {"season": 6, "char_a": "Jon Snow", "char_b": "Ramsay Bolton", "interaction": "battle", "sentiment": -0.9},
    {"season": 8, "char_a": "Jon Snow", "char_b": "Daenerys Targaryen", "interaction": "betrayal", "sentiment": -0.7}
]
df = pd.DataFrame(data)

# Build a dynamic network graph
def build_network(season):
    G = nx.DiGraph()
    season_data = df[df['season'] <= season]
    
    for _, row in season_data.iterrows():
        weight = 1 if row['sentiment'] > 0 else -1
        G.add_edge(row['char_a'], row['char_b'], weight=weight)
    
    pos = nx.spring_layout(G, seed=42)
    positive_edges = [(u, v) for u, v, d in G.edges(data=True) if d['weight'] == 1]
    negative_edges = [(u, v) for u, v, d in G.edges(data=True) if d['weight'] == -1]
    
    nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=3000, font_size=10, edge_color='gray')
    nx.draw_networkx_edges(G, pos, edgelist=positive_edges, edge_color='blue', width=2)
    nx.draw_networkx_edges(G, pos, edgelist=negative_edges, edge_color='red', width=2, style='dashed')
    plt.title(f'Power Network - Season {season}')
    plt.show()

# Sentiment Analysis of Dialogue
def analyze_sentiment(dialogue):
    analyzer = SentimentIntensityAnalyzer()
    sentiment = analyzer.polarity_scores(dialogue)
    return sentiment['compound']

# Example usage
print("Sentiment of 'You are a traitor!':", analyze_sentiment("You are a traitor!"))

# Machine Learning Model for Power Prediction
features = [
    [0.7, 5, 0],  # Ned Stark (Strong alliance, no betrayals)
    [-0.8, 3, 1],  # Ned vs. Cersei (Conflict, betrayal detected)
    [0.5, 4, 0],   # Jaime & Brienne (Trust, no betrayal)
    [-0.9, 2, 1],  # Jon vs. Ramsay (Battle, rivalry)
    [-0.7, 3, 1]   # Jon vs. Daenerys (Betrayal, high tension)
]
target = [0.8, 0.4, 0.7, 0.3, 0.5]  # Next season's power score

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

predictions = model.predict(X_test)
print("Predicted power scores for next season:", predictions)

# Build network for Season 6
build_network(6)
```

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import plotly.graph_objects as go
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

# Sample dataset for character interactions over seasons
data = [
    {"season": 1, "char_a": "Ned Stark", "char_b": "Robert Baratheon", "interaction": "alliance", "sentiment": 0.7},
    {"season": 1, "char_a": "Ned Stark", "char_b": "Cersei Lannister", "interaction": "conflict", "sentiment": -0.8},
    {"season": 3, "char_a": "Jaime Lannister", "char_b": "Brienne of Tarth", "interaction": "trust", "sentiment": 0.5},
    {"season": 6, "char_a": "Jon Snow", "char_b": "Ramsay Bolton", "interaction": "battle", "sentiment": -0.9},
    {"season": 8, "char_a": "Jon Snow", "char_b": "Daenerys Targaryen", "interaction": "betrayal", "sentiment": -0.7}
]
df = pd.DataFrame(data)

# Build a dynamic network graph
def build_network(season):
    G = nx.DiGraph()
    season_data = df[df['season'] <= season]
    
    for _, row in season_data.iterrows():
        weight = 1 if row['sentiment'] > 0 else -1
        G.add_edge(row['char_a'], row['char_b'], weight=weight)
    
    pos = nx.spring_layout(G, seed=42)
    positive_edges = [(u, v) for u, v, d in G.edges(data=True) if d['weight'] == 1]
    negative_edges = [(u, v) for u, v, d in G.edges(data=True) if d['weight'] == -1]
    
    nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=3000, font_size=10, edge_color='gray')
    nx.draw_networkx_edges(G, pos, edgelist=positive_edges, edge_color='blue', width=2)
    nx.draw_networkx_edges(G, pos, edgelist=negative_edges, edge_color='red', width=2, style='dashed')
    plt.title(f'Power Network - Season {season}')
    plt.show()

# Interactive Visualization using Plotly
def plot_interactive_network(season):
    G = nx.DiGraph()
    season_data = df[df['season'] <= season]
    for _, row in season_data.iterrows():
        weight = 1 if row['sentiment'] > 0 else -1
        G.add_edge(row['char_a'], row['char_b'], weight=weight)
    
    edge_x, edge_y, edge_text = [], [], []
    pos = nx.spring_layout(G, seed=42)
    
    for edge in G.edges():
        x0, y0 = pos[edge[0]]
        x1, y1 = pos[edge[1]]
        edge_x.extend([x0, x1, None])
        edge_y.extend([y0, y1, None])
        edge_text.append(f'{edge[0]} → {edge[1]}')
    
    edge_trace = go.Scatter(x=edge_x, y=edge_y, line=dict(width=2, color='gray'), hoverinfo='text', mode='lines', text=edge_text)
    node_x, node_y, node_text = [], [], []
    
    for node in G.nodes():
        x, y = pos[node]
        node_x.append(x)
        node_y.append(y)
        node_text.append(node)
    
    node_trace = go.Scatter(x=node_x, y=node_y, mode='markers+text', text=node_text, textposition='top center', marker=dict(size=10, color='blue'))
    
    fig = go.Figure(data=[edge_trace, node_trace])
    fig.update_layout(title=f'Interactive Power Network - Season {season}', showlegend=False)
    fig.show()

# Sentiment Analysis of Dialogue
def analyze_sentiment(dialogue):
    analyzer = SentimentIntensityAnalyzer()
    sentiment = analyzer.polarity_scores(dialogue)
    return sentiment['compound']

# Example usage
print("Sentiment of 'You are a traitor!':", analyze_sentiment("You are a traitor!"))

# Machine Learning Model for Power Prediction
features = [
    [0.7, 5, 0],  # Ned Stark (Strong alliance, no betrayals)
    [-0.8, 3, 1],  # Ned vs. Cersei (Conflict, betrayal detected)
    [0.5, 4, 0],   # Jaime & Brienne (Trust, no betrayal)
    [-0.9, 2, 1],  # Jon vs. Ramsay (Battle, rivalry)
    [-0.7, 3, 1]   # Jon vs. Daenerys (Betrayal, high tension)
]
target = [0.8, 0.4, 0.7, 0.3, 0.5]  # Next season's power score

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

predictions = model.predict(X_test)
print("Predicted power scores for next season:", predictions)

# Build interactive network for Season 6
plot_interactive_network(6)
```
